```python
import dspy
class SimpleAITeacher(dspy.Signature):
    """Explain AI concepts in simple english."""
    ai_concepts = dspy.InputField()
    simple_english = dspy.OutputField()

```

```python
from google.colab import userdata
OPENAI_API_KEY=userdata.get('OPENAI_API_KEY')
turbo = dspy.OpenAI(model='gpt-3.5-turbo', max_tokens=1000, api_key=OPENAI_API_KEY)
```

```python
dspy.settings.configure(lm=turbo)
from dspy.signatures.signature import signature_to_template
simple_explanation_as_template = signature_to_template(SimpleAITeacher)

```

```python
print(str(simple_explanation_as_template))
```

```python
class CoT(dspy.Module):
    def __init__(self):
        super().__init__()
        self.prog = dspy.ChainOfThought(SimpleAITeacher)
    def forward(self, ai_concepts):
        return self.prog(ai_concepts=ai_concepts)
c = CoT()
```

```python
c.forward("What is Generative AI")
```
